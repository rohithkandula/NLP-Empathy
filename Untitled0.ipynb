```python
# Cell 1 — Setup
!pip -q install --upgrade "tensorflow==2.19.0" scikit-learn==1.6.1 pandas==2.2.2 matplotlib==3.9.2 || true

import os, re, unicodedata, random, math, json, time
from pathlib import Path
import numpy as np, pandas as pd, matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers, models, callbacks, optimizers
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.preprocessing.text import tokenizer_from_json
from sklearn.metrics import mean_absolute_error, mean_squared_error
from scipy import stats
from sklearn.model_selection import train_test_split

RANDOM_SEED = 42
np.random.seed(RANDOM_SEED); tf.random.set_seed(RANDOM_SEED); random.seed(RANDOM_SEED)

ART_DIR = "/content/cw2_artifacts"
os.makedirs(ART_DIR, exist_ok=True)
print("TF:", tf.__version__, "| ART_DIR:", ART_DIR)
```

```python
# Cell 2 — Upload or use existing CSV. Will create /content/wassa_conv_train_all.csv
DATA_CSV = "/content/wassa_conv_train_all.csv"

if not os.path.exists(DATA_CSV):
    from google.colab import files
    print("Please upload the WASSA train TSV (e.g. WASSA23_conv_level_with_labels_train.tsv)")
    uploaded = files.upload()
    tsv = None
    for name in uploaded.keys():
        if name.lower().endswith(".tsv") or name.lower().endswith(".csv"):
            tsv = name; break
    assert tsv is not None, "No TSV/CSV uploaded."
    print("Loaded:", tsv)
    df_raw = pd.read_csv(tsv, sep="\t", dtype=str, engine="python")
    # pick label heuristically (Empathy / Emotion / EmotionalPolarity)
    label_candidates = [c for c in df_raw.columns if c.lower() in ("empathy","emotion","emotionalpolarity","emotional_polarity")]
    assert label_candidates, f"No label column found in uploaded file; columns: {list(df_raw.columns)}"
    label_col = label_candidates[0]
    df = df_raw[["text", label_col]].rename(columns={label_col:"label"})
    df = df.dropna(subset=["text","label"])
    # ensure numeric labels if possible
    df = df[df["label"].astype(str).str.match(r"^-?\d+(\.\d+)?$")]
    df["label"] = df["label"].astype(float)
    df.to_csv(DATA_CSV, index=False)
    print("Saved cleaned CSV to", DATA_CSV, "rows:", len(df))
else:
    df = pd.read_csv(DATA_CSV)
    print("Using existing CSV:", DATA_CSV, "rows:", len(df))

display(df.head())
print("Label range:", df['label'].min(), "to", df['label'].max())
```

```python
# Cell 3 — text cleaning and handcrafted features
EMOJI_PATTERN   = re.compile("[" "\U0001F600-\U0001F64F" "\U0001F300-\U0001F5FF" "\U0001F680-\U0001F6FF" "\U0001F1E0-\U0001F1FF" "]+")
URL_PATTERN     = re.compile(r"(https?://\S+|www\.\S+)")
MENTION_PATTERN = re.compile(r"@\w+")
HASHTAG_PATTERN = re.compile(r"#\w+")
ELONG_PATTERN   = re.compile(r"(.)\1{2,}")

def basic_clean(s):
    if not isinstance(s, str): return ""
    s = s.strip()
    s = unicodedata.normalize("NFKC", s)
    s = URL_PATTERN.sub(" <URL> ", s)
    s = MENTION_PATTERN.sub(" <USER> ", s)
    s = HASHTAG_PATTERN.sub(" <HASHTAG> ", s)
    s = ELONG_PATTERN.sub(r"\1\1\1", s)
    s = s.lower()
    return s

def handcrafted_features(text):
    t = text if isinstance(text,str) else ""
    ex = t.count("!")
    qn = t.count("?")
    toks = t.split()
    caps = sum(1 for w in toks if w.isupper() and len(w)>1)
    allcaps_ratio = caps / max(1, len(toks))
    emoji_count = len(EMOJI_PATTERN.findall(t))
    ell = t.count("...")
    elong = sum(1 for w in toks if ELONG_PATTERN.search(w))
    sarcasm = int(bool(re.search(r"\b(yeah right|sure|as if|right then)\b", t)))
    negation = int(bool(re.search(r"\b(not|n't|never|no way|nothing)\b", t)))
    return np.array([ex, qn, allcaps_ratio, emoji_count, ell, elong, sarcasm, negation], dtype=float)

df["clean_text"] = df["text"].astype(str).map(basic_clean)
HF = np.vstack(df["clean_text"].map(handcrafted_features).values)
print("HF shape:", HF.shape, "| sample features (first row):", HF[0])
```

```python
# Cell 4 — tokenizer robust load/create + build sequences
MAX_VOCAB = 20000
MAX_LEN = 80
tokenizer_json_path = os.path.join(ART_DIR, "tokenizer.json")

def create_and_save_tokenizer(texts, path, num_words=MAX_VOCAB):
    tok = Tokenizer(num_words=num_words, oov_token="<OOV>")
    tok.fit_on_texts(texts)
    open(path, "w").write(tok.to_json())
    print("[created] tokenizer saved to:", path)
    return tok

tokenizer = None
if os.path.exists(tokenizer_json_path):
    try:
        raw = open(tokenizer_json_path, "r").read()
        tokenizer = tokenizer_from_json(raw)
        print("[loaded] tokenizer_from_json OK")
    except Exception as e1:
        print("[warning] tokenizer_from_json failed:", e1)
        try:
            obj = json.load(open(tokenizer_json_path, "r"))
            if isinstance(obj, dict) and "word_index" in obj:
                tok = Tokenizer(num_words=MAX_VOCAB, oov_token="<OOV>")
                tok.word_index = obj["word_index"]
                open(tokenizer_json_path, "w").write(tok.to_json())
                tokenizer = tok
                print("[reconstructed] tokenizer from word_index and re-saved")
            else:
                print("[info] tokenizer.json exists but not in expected word_index format")
        except Exception as e2:
            print("[warning] parsing tokenizer.json failed:", e2)

if tokenizer is None:
    tokenizer = create_and_save_tokenizer(df["clean_text"].tolist(), tokenizer_json_path, num_words=MAX_VOCAB)

sequences = tokenizer.texts_to_sequences(df["clean_text"].tolist())
X_seq = pad_sequences(sequences, maxlen=MAX_LEN, padding="post", truncating="post")
# prev-turn sequences (simple shift method)
prev_texts = [""] + df["clean_text"].tolist()[:-1]
prev_seq = tokenizer.texts_to_sequences(prev_texts)
X_prev = pad_sequences(prev_seq, maxlen=MAX_LEN, padding="post", truncating="post")
print("X_seq shape:", X_seq.shape, "X_prev shape:", X_prev.shape, "vocab:", len(tokenizer.word_index))
```

```python
# Cell 5 — index-based splits for reproducibility
indices = np.arange(len(df))
train_idx, test_idx = train_test_split(indices, test_size=0.15, random_state=RANDOM_SEED)
train_idx, val_idx = train_test_split(train_idx, test_size=0.1765, random_state=RANDOM_SEED)  # ~0.15 val of full

X_train_seq = X_seq[train_idx]; X_val_seq = X_seq[val_idx]; X_test_seq = X_seq[test_idx]
X_train_prev = X_prev[train_idx]; X_val_prev = X_prev[val_idx]; X_test_prev = X_prev[test_idx]
HF_train = HF[train_idx]; HF_val = HF[val_idx]; HF_test = HF[test_idx]
y = df["label"].values
y_train = y[train_idx]; y_val = y[val_idx]; y_test = y[test_idx]

print("Train/Val/Test sizes:", len(train_idx), len(val_idx), len(test_idx))
```

```python
# Cell 6 — build model with named attention weights for extraction
def build_model(vocab_size=len(tokenizer.word_index)+1, embedding_dim=100, max_len=MAX_LEN,
                use_prev=True, use_handcrafted=True, use_attention=True):
    input_seq = layers.Input(shape=(max_len,), name="input_seq")
    input_prev = layers.Input(shape=(max_len,), name="input_prev") if use_prev else None
    input_hf = layers.Input(shape=(HF.shape[1],), name="input_hf") if use_handcrafted else None

    emb = layers.Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=max_len, name="emb")(input_seq)
    lstm = layers.Bidirectional(layers.LSTM(128, return_sequences=True))(emb)

    if use_attention:
        attn_score = layers.Dense(1, name="attn_score")(lstm)
        attn_flat = layers.Flatten()(attn_score)
        attn_weights = layers.Activation('softmax', name="attn_weights")(attn_flat)
        attn_reshaped = layers.Reshape((max_len,1))(attn_weights)
        weighted = layers.multiply([lstm, attn_reshaped])
        context = layers.Lambda(lambda x: tf.reduce_sum(x, axis=1), name="attn_context")(weighted)
    else:
        context = layers.GlobalAveragePooling1D()(lstm)

    concat_list = [context]
    if use_prev:
        emb_prev = layers.Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=max_len, name="emb_prev")(input_prev)
        lstm_prev = layers.Bidirectional(layers.LSTM(64, return_sequences=False))(emb_prev)
        concat_list.append(lstm_prev)
    if use_handcrafted:
        concat_list.append(input_hf)

    x = layers.concatenate(concat_list)
    x = layers.Dense(128, activation="relu")(x)
    x = layers.Dropout(0.3)(x)
    out = layers.Dense(1, activation="linear")(x)

    inputs = [input_seq]
    if use_prev: inputs.append(input_prev)
    if use_handcrafted: inputs.append(input_hf)

    model = models.Model(inputs=inputs, outputs=out)
    model.compile(optimizer=optimizers.Adam(learning_rate=5e-4), loss="mse")
    return model

model = build_model()
model.summary()
```

```python
# Cell 7 — train helper and train (saves model, preds, curves)
def compile_and_train(use_prev=True, use_handcrafted=True, use_attention=True, name="model"):
    model = build_model(use_prev=use_prev, use_handcrafted=use_handcrafted, use_attention=use_attention)
    early = callbacks.EarlyStopping(patience=6, restore_best_weights=True, monitor="val_loss", verbose=1)
    checkpoint = callbacks.ModelCheckpoint(os.path.join(ART_DIR, f"{name}_best.keras"), save_best_only=True, monitor="val_loss", verbose=1)
    reduce_lr = callbacks.ReduceLROnPlateau(monitor="val_loss", factor=0.5, patience=3, min_lr=1e-6, verbose=1)
    cb_list = [early, checkpoint, reduce_lr]

    train_inputs = [X_train_seq]
    val_inputs = [X_val_seq]
    if use_prev:
        train_inputs.append(X_train_prev); val_inputs.append(X_val_prev)
    if use_handcrafted:
        train_inputs.append(HF_train); val_inputs.append(HF_val)

    history = model.fit(
        x=train_inputs, y=y_train,
        validation_data=(val_inputs, y_val),
        epochs=50, batch_size=64,
        callbacks=cb_list, verbose=2
    )

    # save curves
    plt.figure(figsize=(6,4))
    plt.plot(history.history["loss"], label="train")
    plt.plot(history.history["val_loss"], label="val")
    plt.legend(); plt.xlabel("epoch"); plt.ylabel("MSE")
    plt.title(name + " training curve")
    plt.savefig(os.path.join(ART_DIR, f"{name}_train_curve.png"), bbox_inches="tight")
    plt.close()

    # evaluate on test
    test_inputs = [X_test_seq]
    if use_prev: test_inputs.append(X_test_prev)
    if use_handcrafted: test_inputs.append(HF_test)

    preds = model.predict(test_inputs).ravel()
    mae = mean_absolute_error(y_test, preds)
    rmse = math.sqrt(mean_squared_error(y_test, preds))
    pear = float(stats.pearsonr(y_test, preds)[0]) if len(set(y_test))>1 else 0.0

    # save predictions aligned using test_idx
    out_df = pd.DataFrame({
        "text": df.loc[test_idx, "text"].values,
        "y_true": y_test,
        "y_pred": preds
    })
    out_df.to_csv(os.path.join(ART_DIR, f"{name}_test_predictions.csv"), index=False)
    model.save(os.path.join(ART_DIR, f"{name}_final.keras"))

    metrics = {"MAE":mae, "RMSE":rmse, "Pearson":pear}
    print(name, " -> ", metrics)
    return model, metrics, preds

# Run full hybrid training
hybrid_model, hybrid_metrics, hybrid_preds = compile_and_train(use_prev=True, use_handcrafted=True, use_attention=True, name="hybrid_bilstm_attn")
```

```python
# Cell 8 — ablation experiments
models_metrics = {"hybrid": hybrid_metrics}

m_no_hf, m_no_hf_metrics, _ = compile_and_train(use_prev=True, use_handcrafted=False, use_attention=True, name="no_hf")
models_metrics["no_hf"] = m_no_hf_metrics

m_no_prev, m_no_prev_metrics, _ = compile_and_train(use_prev=False, use_handcrafted=True, use_attention=True, name="no_prev")
models_metrics["no_prev"] = m_no_prev_metrics

m_no_attn, m_no_attn_metrics, _ = compile_and_train(use_prev=True, use_handcrafted=True, use_attention=False, name="no_attn")
models_metrics["no_attn"] = m_no_attn_metrics

rows=[]
for k,v in models_metrics.items():
    rows.append({"model":k, "MAE":v["MAE"], "RMSE":v["RMSE"], "Pearson":v["Pearson"]})
comp_df = pd.DataFrame(rows)
comp_df.to_csv(os.path.join(ART_DIR, "cw2_model_comparison.csv"), index=False)
print("Saved comparison:", os.path.join(ART_DIR, "cw2_model_comparison.csv"))
display(comp_df)
```
